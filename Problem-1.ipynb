```python
# Utility Function to analyze a DataFrame

def analyze(df):
    # Print the number of rows in the DataFrame
    print(f"Number of rows: {len(df)}")
    
    # Print the number of columns in the DataFrame
    print(f"Number of columns: {len(df.columns)}")
    
    # Create a DataFrame with information about the original DataFrame
    # NUNIQUE: Number of unique values in each column
    # DTYPE: Data type of each column
    # NaN Values: Number of missing (NaN) values in each column
    info_df = pd.DataFrame({
        "NUNIQUE": df.nunique(),
        "DTYPE": df.dtypes,
        "NaN Values": df.isna().sum()
    })
    
    # Print a blank line for better readability
    print()
    
    # Print the info DataFrame
    print(info_df)
```

```python
# Merging two DataFrames on the 'CUSTOMER_ID' column
# Left join ensures all rows from 'pack_purc_df' are kept, with matching rows from 'cus_prof_df'

merged_df = pd.merge(
    pack_purc_df,  # The left DataFrame to merge
    cus_prof_df,   # The right DataFrame to merge
    on=['CUSTOMER_ID'],  # Column name(s) to join on
    how='left'     # Type of merge to perform ('left' keeps all rows from the left DataFrame)
)

# 'merged_df' now contains all columns from 'pack_purc_df' and matching columns from 'cus_prof_df'
# Rows in 'pack_purc_df' without matching 'CUSTOMER_ID' in 'cus_prof_df' will have NaN for 'cus_prof_df' columns
```

```python
analyze(merged_df)
```

```python
# Merging the 'prod_cat_df' DataFrame with the previously merged DataFrame 'merged_df'
# The merge is performed on the 'PACK_ID' and 'SERVICE_TYPE' columns

merged_df1 = pd.merge(
    prod_cat_df,  # The left DataFrame to merge
    merged_df,    # The right DataFrame to merge
    on=['PACK_ID', 'SERVICE_TYPE']  # Column names to join on
)

# 'merged_df1' now contains all columns from 'prod_cat_df' and matching columns from 'merged_df'
# Only rows with matching 'PACK_ID' and 'SERVICE_TYPE' in both DataFrames will be included
```

```python
analyze(merged_df1)
```

```python
# Initialize a LabelEncoder instance
le = LabelEncoder()

# Apply the LabelEncoder to the 'SERVICE_TYPE' column
# This transforms categorical values in 'SERVICE_TYPE' to numerical values
merged_df1['SERVICE_TYPE'] = le.fit_transform(merged_df1['SERVICE_TYPE'])

# Apply the LabelEncoder to the 'CONNECTION_TYPE' column
# This transforms categorical values in 'CONNECTION_TYPE' to numerical values
merged_df1['CONNECTION_TYPE'] = le.fit_transform(merged_df1['CONNECTION_TYPE'])

# Apply the LabelEncoder to the 'PRODUCT_STATUS' column
# This transforms categorical values in 'PRODUCT_STATUS' to numerical values
merged_df1['PRODUCT_STATUS'] = le.fit_transform(merged_df1['PRODUCT_STATUS'])

# Apply the LabelEncoder to the 'PAYMENT_METHOD' column
# This transforms categorical values in 'PAYMENT_METHOD' to numerical values
merged_df1['PAYMENT_METHOD'] = le.fit_transform(merged_df1['PAYMENT_METHOD'])

# Apply the LabelEncoder to the 'ROUTER_CATEGORY' column
# This transforms categorical values in 'ROUTER_CATEGORY' to numerical values
merged_df1['ROUTER_CATEGORY'] = le.fit_transform(merged_df1['ROUTER_CATEGORY'])
```

```python
analyze(merged_df1)
```

```python
# Import the train_test_split function from sklearn.model_selection
# This function is used to split the data into training and testing sets
from sklearn.model_selection import train_test_split

# Import the RandomForestClassifier from sklearn.ensemble
# This is a machine learning model used for classification tasks
from sklearn.ensemble import RandomForestClassifier

# Import the accuracy_score function from sklearn.metrics
# This function is used to evaluate the accuracy of the classification model
from sklearn.metrics import accuracy_score
```

```python
analyze(problem_df)
```

```python
# Merge 'problem_df' with 'merged_df1' on the 'CUSTOMER_ID' column
# Left join ensures all rows from 'problem_df' are kept, with matching rows from 'merged_df1'

problem_data = pd.merge(
    problem_df,    # The left DataFrame to merge
    merged_df1,    # The right DataFrame to merge
    on='CUSTOMER_ID',  # Column name to join on
    how='left'     # Type of merge to perform ('left' keeps all rows from 'problem_df')
)

# 'problem_data' now contains all columns from 'problem_df' and matching columns from 'merged_df1'
# Rows in 'problem_df' without matching 'CUSTOMER_ID' in 'merged_df1' will have NaN for 'merged_df1' columns
```

```python
analyze(problem_data)
```

```python
# Replace NaN values in the 'PACK_ID' column with -1 for rows that contain any NaN values in the DataFrame

# Locate rows in 'problem_data' where any column has NaN values
# For these rows, set the 'PACK_ID' column value to -1
problem_data.loc[problem_data.isnull().any(axis=1), 'PACK_ID'] = -1
```

```python
analyze(problem_data)
```

```python
# Drop duplicate rows based on the 'CUSTOMER_ID' column

# The 'drop_duplicates' function removes duplicate rows from the DataFrame
# The 'subset' parameter specifies the column(s) to consider when identifying duplicates
# In this case, duplicates are identified based on the 'CUSTOMER_ID' column
problem_data = problem_data.drop_duplicates(subset=['CUSTOMER_ID'])
```

```python
analyze(problem_data)
```

```python
# Define the feature matrix 'X' by dropping specified columns from 'problem_data'
# These columns are dropped because they have no correlation with the target variable 'PACK_ID'

X = problem_data.drop([
    'CUSTOMER_ID',                   # Unique identifier for customers, not a feature
    'PACK_ID',                       # Target variable, should not be in the feature set
    'HITS',                          # No correlation with 'PACK_ID'
    'STREAMING_VOL_PERCENTAGE',      # No correlation with 'PACK_ID'
    'WEB_BROWSING_VOL_PERCENTAGE',   # No correlation with 'PACK_ID'
    'IM_VOICE_CALL_VOL_PERCENTAGE',  # No correlation with 'PACK_ID'
    'SNS_VOL_PERCENTAGE',            # No correlation with 'PACK_ID'
    'FILE_ACCESS_VOL_PERCENTAGE',    # No correlation with 'PACK_ID'
    'NAVIGATION_VOL_PERCENTAGE',     # No correlation with 'PACK_ID'
    'OTHERS_VOL_PERCENTAGE',         # No correlation with 'PACK_ID'
    'GAME_VOL_PERCENTAGE',           # No correlation with 'PACK_ID'
    'EMAIL_VOL_PERCENTAGE',          # No correlation with 'PACK_ID'
], axis=1)

# Define the target variable 'y' as the 'PACK_ID' column from 'problem_data'
y = problem_data['PACK_ID']
```

```python
# Split the dataset into training and testing sets
# The dataset is split such that 30% of the data is used for testing and 70% for training
# 'random_state=42' ensures reproducibility of the split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

rf_classifier = RandomForestClassifier()

# Train the Random Forest classifier on the training data
rf_classifier.fit(X_train, y_train)
```

```python
# Step 4: Model Evaluation
# Predict on test set
y_pred = rf_classifier.predict(X_test)

# Evaluate model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy*100)
```

```python
# Use the trained Random Forest classifier to make predictions on the entire dataset 'X'
# This is typically done on the test set, but here it's being applied to the entire dataset

predictions = rf_classifier.predict(X)

# 'predictions' now contains the predicted 'PACK_ID' for each row in 'X'
```

```python
# Create a DataFrame for submission with 'CUSTOMER_ID' from 'problem_df' and the predicted 'PACK_ID'

submission = pd.DataFrame({
    'CUSTOMER_ID': problem_df['CUSTOMER_ID'],  # Use 'CUSTOMER_ID' from 'problem_df'
    'PACK_ID': [pred for pred in predictions]  # Add the predictions for 'PACK_ID'
})

# The 'submission' DataFrame now contains 'CUSTOMER_ID' and the corresponding predicted 'PACK_ID'
```

```python
### write CSV file to S3
write_file_name = 'Q1-submission.csv'

# Convert the dataframe to CSV
csv_buffer = StringIO()
submission.to_csv(csv_buffer, index = False)

# Write the CSV data to S3
s3_client.put_object(Bucket=bucket_name, Key=write_file_name, Body=csv_buffer.getvalue())
```
